```python
values = [1,2,3,4,5,6]
alias = values
alias.append(7)
print("Values: ",  values)
print("Alias: ",  alias)
```

```python
values = [1,2,3,4,5,6]
alias = values.copy()
alias.append(7)
print("Values: ",  values)
print("Alias: ",  alias)
```

```python
record = {"Name" : "Kamal",
     "Age" : 19,
     "classes" : ["Math", "Programming", "Statistic"]}
record_alias = record
record_alias["Age"] = 21
print("record: ",  record)
print("record_alias: ",  record_alias)
```

```python
record = {"Name" : "Kamal",
     "Age" : 19,
     "classes" : ["Math", "Programming", "Statistic"]}
record_copy = record.copy()
record_copy["Age"] = 21
print("record: ",  record)
print("record_copy : ",  record_copy)
```

When I assigned the list and dictionary to an alias, 
both names pointed to the same object in memory. 
Because of that, changing the alias also changed the original variable. 
When I created a copy using .copy() or list(), a new object was created, so changes to the copy did not affect the original.

```python
events = [
    {"user_id": 1, "event_type": "login", "duration_seconds": 120},
    {"user_id": 2, "event_type": "logout", "duration_seconds": 300},
    {"user_id": 3, "event_type": "purchase", "duration_seconds": 450},
    {"user_id": 4, "event_type": "login", "duration_seconds": -50},     
    {"user_id": 5, "event_type": "login", "duration_seconds": 200},
    {"user_id": 6, "event_type": "logout", "duration_seconds": "Nan"},   
    {"user_id": 7, "event_type": "purchase", "duration_seconds": 600},
    {"user_id": 8, "event_type": "login", "duration_seconds": "Nan"},
    {"user_id": 9, "event_type": "logout", "duration_seconds": 240},
    {"user_id": 10, "event_type": "purchase", "duration_seconds": 360},
    {"user_id": 11, "event_type": "login", "duration_seconds": 150},
    {"user_id": 12, "event_type": "logout", "duration_seconds": 90}
]


cleaned_events = []

for event in events:
    duration = event.get("duration_seconds")
    if isinstance(duration, (int, float)) and duration > 0:
        event["duration_minutes"] = duration / 60
        cleaned_events.append(event)


print(f"Total valid events: {len(cleaned_events)}")
print("First cleaned event:", cleaned_events[0])
print("Second cleaned event:", cleaned_events[1])
```

```python
event_counts = {}
for event in cleaned_events:
    etype = event["event_type"]
    event_counts[etype] = event_counts.get(etype, 0) + 1


event_sums = {}
event_nums = {}

for event in cleaned_events:
    etype = event["event_type"]
    duration = event["duration_minutes"]
    event_sums[etype] = event_sums.get(etype, 0) + duration
    event_nums[etype] = event_nums.get(etype, 0) + 1

event_avg_duration = {etype: event_sums[etype]/event_nums[etype] for etype in event_sums}

unique_users = len(set([event["user_id"] for event in cleaned_events]))
```

```python
print("Sum of event counts matches cleaned_events:", sum(event_counts.values()) == len(cleaned_events))
```

```python
csv_rows = ["metric,key,value"]

for etype, count in event_counts.items():
    csv_rows.append(f"count,{etype},{count}")

for etype, avg in event_avg_duration.items():
    csv_rows.append(f"average,{etype},{avg:.2f}")

csv_rows.append(f"unique_users,all,{unique_users}")

csv_string = "\n".join(csv_rows)

print("\nCSV output:\n")
print(csv_string)
```
